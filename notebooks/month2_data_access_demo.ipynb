# ScrollGuard Month 2 Data Access Demo

This notebook demonstrates the Month 2 data-access workflow:

- load metadata
- verify cached Vesuvius CT crop
- inspect crop statistics
- display the metadata index

This demo uses the local crop cache and does not contact the Vesuvius server.


```python
from pathlib import Path
import json
import numpy as np

cache_path = Path('../data_cache/crops/Scroll1_z1000_y3520_3584_x4256_4320.npy')
metadata_path = Path('../outputs/cached_tiny_crop_metadata.json')
index_path = Path('../outputs/metadata_index.csv')

print(cache_path.exists(), metadata_path.exists(), index_path.exists())
```

```python
crop = np.load(cache_path)
metadata = json.loads(metadata_path.read_text())

print('shape:', crop.shape)
print('dtype:', crop.dtype)
print('min:', crop.min())
print('max:', crop.max())
print('mean:', crop.mean())
print('metadata:', metadata)
```
